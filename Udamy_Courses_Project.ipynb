```python
import pandas as pd
df1=pd.DataFrame(pd.read_csv('Entry Level Project-web-development.csv'))
df2=pd.DataFrame(pd.read_csv('business-courses.csv'))
df3=pd.DataFrame(pd.read_csv('design-courses.csv'))
df4=pd.DataFrame(pd.read_csv('music-courses.csv'))
df5=pd.DataFrame(pd.read_csv('web-development.csv'))
df=pd.concat([df1,df2,df3,df4,df5])
df.head()
```

```python
df.dtypes
```

```python
df['course_id'].fillna(0, inplace=True)
df['url'].fillna(0, inplace=True)
df['price'].fillna(0, inplace=True)
df['published_timestamp'].fillna(0, inplace=True)
df['level'].fillna(0, inplace=True)
df['subject'].fillna(0, inplace=True)
df['Free/Paid'].fillna(0, inplace=True)
df['Date'].fillna(0, inplace=True)
df['rating'].fillna(0, inplace=True)
df['content_duration'].fillna(0, inplace=True)
df['num_subscribers'].fillna(0, inplace=True)
df['num_reviews'].fillna(0, inplace=True)
df['num_lectures'].fillna(0, inplace=True)
df['course_title'].fillna(0, inplace=True)
df.isnull().sum()
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

Repaire data **types**

```python

df['course_id']=df['course_id'].astype('string')
df['num_subscribers']=df['num_subscribers'].astype('int64')
df['num_reviews']=df['num_reviews'].astype('int64')
df['num_lectures']=df['num_lectures'].astype('int64')
df['published_timestamp']=pd.to_datetime(df['published_timestamp'],errors='coerce')
df['course_title']=df['course_title'].astype('string')
df['level']=df['level'].astype('string')
df['subject']=df['subject'].astype('string')
df['Free/Paid']=df['Free/Paid'].astype('string')
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
df.dtypes
```

```python
df['year'] = df['Date'].dt.year
df['year'] = df['year'].fillna(0).astype('int64')
df['year']
```

```python
df = df.drop('Rating', axis=1)
```

# The most trand course at last 5 years

```python
for year in [2013,2014,2015,2016,2017]:
    print('the most tranding course in ', year,df[df['year']==year].groupby('subject')['num_subscribers'].sum().sort_values(ascending=False).head(1))
```

# The prices average for eash course supject and for eash year

```python
for year in [2013,2014,2015,2016,2017]:
      print('the avarage price of courses in ', year,df[df['year']==year].groupby('subject')['price'].mean())
```

# The titles and number of subscription for each course subject

```python

for year in [2013,2014,2015, 2016, 2017]:
    print('the titles and num of subscribers for each course subject in ', year,
          df[df['year'] == year].groupby(['subject', 'course_title'])['num_subscribers'].sum())
```

# The avarage price of each course supject level

```python
for year in [2013,2014,2015, 2016, 2017]:
      print('the avarage price of each course supject level', year,
                df[df['year'] == year].groupby(['subject', 'level'])['price'].mean())
```

# The number of subscribers of free and paid course at last 5 years

```python
for year in [2013,2014,2015, 2016, 2017]:
      print('the free course subject and title has maximum number of subscription  ', year,
                df[df['year'] == year].groupby(['subject', 'Free/Paid'])['num_subscribers'].sum())
```

# Top free courses at last 5 years

```python
for year in [2013,2014,2015, 2016, 2017]:
     df_year = df[df['year'] == year]
     result = df_year[df_year['Free/Paid'] == 'Free'].groupby(['subject', 'course_title'])['num_subscribers'].max()
     print('the titles and num of subscribers for each course subject in ', year,result)
```

# Top paid courses at last 5 years

```python
for year in [2013,2014,2015, 2016, 2017]:
       df_year = df[df['year'] == year]
       result1 = df_year[df_year['Free/Paid'] == 'Paid'].groupby(['subject', 'course_title'])['num_subscribers'].max()
       print('the titles and num of subscribers for each course subject in ', year,result1)
```

# Top 5 most expensive courses

```python
df_price=df.groupby(['subject','course_title'])['price'].max().sort_values(ascending=False).head(5)
print('Top5 expensive courses are',df_price)
```

# The total number off subscripers at every course level

```python
df_level=df.groupby(df['level'])['num_subscribers'].sum()
print('The total number off subscripers at every course level is ',df_level)
```

# Top course in number of subscription and rating and number of reviews

```python
df_top_course = (df.groupby(['subject','course_title'])['num_subscribers'].max().to_frame().reset_index()).merge(df.groupby(['subject','course_title'])['rating'].max().to_frame().reset_index()).merge(df.groupby(['subject','course_title'])['num_reviews'].max().to_frame().reset_index())
print('Top courses at udamy academy are ',df_top_course)
```
